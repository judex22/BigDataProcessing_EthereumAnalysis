# Number of Transactions per month

```python
### PART A ( No. of Transaction per Month)
import pandas as pd
import string
import re
```

```python
filename = 'partAoutput.txt'       # opening txt files for Number of Transactions per month
file = open(filename, 'rt')
text = file.read()
file.close()
# split into words by white space
words = text.split()

t1 = str.maketrans('', '', string.punctuation)
stripped = [w.translate(t1) for w in words]

d=[]
transaction=[]

for i in range(len(stripped)):
    if i%2==0:
        d.append(stripped[i])
    else:
        transaction.append(stripped[i])
        
date=[]

for j in range(len(d)):
    date.append(d[j][:2] + '-' + d[j][2:])

print(date)
print("\n",transaction)
```

```python
import pandas as pd

df = pd.DataFrame(list(zip(date,transaction)),columns=['Date','#Transactions'])
df[['Month','Year']] = df['Date'].str.split('-',expand=True)

```

```python
print(df['Year'].unique())
```

```python
df_final=df.sort_values("Year").reset_index()
df_final

mapping = {"01" : "jan",
                     "02" : "feb",
                     "03" : "mar",
                     "04" : "apr",
                     "05" : "may",
                     "06" : "jun",
                     "07" : "jul",
                     "08" : "aug",
                     "09" : "sept",
                     "10" : "oct",
                     "11" : "nov",
                     "12" : "dec"}

df_final["month"] = df["Month"].map(mapping)
print(df_final)
```

```python
df_15=df_final[0:5]#2015
df_16=df_final[5:17] #2016
df_17=df_final[17:29] #2017
df_18=df_final[29:41] #2018
df_19=df_final[41:]   # 2019
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(5,figsize=(15, 25))


ax[0].bar(df_15['month'], df_15['#Transactions'])
ax[0].set_title('2015')
ax[0].set_ylabel('No.of Transactions')

ax[1].bar(df_16['month'], df_16['#Transactions'])
ax[1].set_title('2016')
ax[1].set_ylabel('No.of Transactions')

ax[2].bar(df_17['month'], df_17['#Transactions'])
ax[2].set_title('2017')
ax[2].set_ylabel('No.of Transactions')

ax[3].bar(df_18['month'], df_18['#Transactions'])
ax[3].set_title('2018')
ax[3].set_ylabel('No.of Transactions')


ax[4].bar(df_19['month'], df_19['#Transactions'])
ax[4].set_title('2019')
ax[4].set_ylabel('No.of Transactions')
```

#  Average Transactions per month

```python
filename_p2 = 'partA_P2.txt'                     #opening txt files for Average Transactions per month
file_p2 = open(filename_p2, 'rt')
text_p2 = file_p2.read()
file_p2.close()
# split into words by white space
words_p2 = text_p2.split()
print(words_p2[0:3])

t2 = str.maketrans('', '', string.punctuation)
s = [x.translate(t2) for x in words_p2]
print(s[0:3],"\n")

date_new=[]
avg_transaction=[]
date_f=[]
date_final=[]

for j in range(len(s)):
    date_new.append(s[j][:6] + ' ' + s[j][6:])
    
            
for z in range(len(date_new)):
    date_f.append(date_new[z][:2] + ' ' + date_new[z][2:])

#for i in range(len(date_f)):
#    if i%2==0:
#        date_final.append(date_f[i])
#    else:
#        avg_transaction.append(date_f[i])

for k in range(len(date_f)):
    date_final.append(date_f[k][:9] + '.' + date_f[k][9:])

        
print(date_final)
#print("\n",avg_transaction)
        
```

```python
import pandas as pd

avg_df = pd.DataFrame(list(zip(date_final)),columns=['Date'])
avg_df[['Month','Year','Avg_Transaction']] = avg_df['Date'].str.split(' ',expand=True)
avg_df = avg_df.drop(columns=['Date'])
avg_df.head(5)
```
